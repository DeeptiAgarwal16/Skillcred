```python
tc_hc=pd.merge(tc[['TransportID', 'StorageID' , 'FuelCosts', 'VehicleMaintenance',
       'DriverWages', 'Tolls', 'DistanceTravelled' , 'VehicleType']],hc[['TransportID',
       'LoadingUnloadingFees', 'LabourCosts', 'EquipmentExpenses',
       'HandlingVolume', 'HandlingLocation']],on="TransportID",how="inner")
tc_hc.head()
```

```python
tc_hc['TotalTransportationCost']=tc_hc.iloc[:,1:4].sum(axis=1)
tc_hc.head()
```

```python
tc_hc['TotalHandlingCost']=tc_hc.iloc[:,8:10].sum(axis=1)
tc_hc.head()
```

```python
tc_hc['TransportationCostPerUnit']=tc_hc['TotalTransportationCost']/tc_hc['HandlingVolume']
tc_hc['HandlingCostPerUnit']=tc_hc['TotalHandlingCost']/tc_hc['HandlingVolume']
tc_hc[['TransportID','TotalTransportationCost','TotalHandlingCost','HandlingVolume','TransportationCostPerUnit','HandlingCostPerUnit']].head()
```

```python
tc_hc_sc=pd.merge(tc_hc,sc[['StorageID', 'RentalFees', 'MaintenanceExpenses',
       'UtilityCosts', 'StorageVolume', 'StorageLocation',
       'Average_Storage_Volume', 'Maximum_Storage_Volume']],on='StorageID',how='inner')
tc_hc_sc.head()
```

```python
tc_hc_sc['StorageUtilisation']=tc_hc_sc['StorageVolume']/tc_hc_sc['Maximum_Storage_Volume']*100
tc_hc_sc['TransportationEfficiency'] = tc_hc_sc['TotalTransportationCost'] / tc_hc_sc['DistanceTravelled']
tc_hc_sc['HandlingEfficiency'] = tc_hc_sc['TotalHandlingCost'] / tc_hc_sc['HandlingVolume']
tc_hc_sc.head()
```

```python
efficiency_metrics = tc_hc_sc[['StorageUtilisation', 'TransportationEfficiency', 'HandlingEfficiency']].mean()
efficiency_metrics
```

```python
from sklearn.cluster import KMeans
clustering_data = tc_hc_sc[['StorageUtilisation', 'TransportationEfficiency', 'HandlingEfficiency']]
kmeans = KMeans(n_clusters=3, random_state=0).fit(clustering_data)
tc_hc_sc['Cluster'] = kmeans.labels_
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.pairplot(tc_hc_sc[['StorageUtilisation', 'TransportationEfficiency', 'HandlingEfficiency', 'Cluster']], hue='Cluster')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 9))
ax = fig.add_subplot(111, projection='3d')
for cluster in tc_hc_sc['Cluster'].unique():
    ax.scatter(tc_hc_sc.loc[tc_hc_sc['Cluster'] == cluster, 'StorageUtilisation'],
               tc_hc_sc.loc[tc_hc_sc['Cluster'] == cluster, 'TransportationEfficiency'],
               tc_hc_sc.loc[tc_hc_sc['Cluster'] == cluster, 'HandlingEfficiency'],
               label=f'Cluster {cluster}')
ax.set_xlabel('Storage Utilisation')
ax.set_ylabel('Transportation Efficiency')
ax.set_zlabel('Handling Efficiency')
ax.legend()
plt.show()
```
